# Question 1: What is the meaning of six sigma in statistics? Give a proper example.

Six Sigma is a statistical measure that represents a process's standard deviation from its mean. It is often used in quality control to measure the variability of a process and identify defects. The goal of Six Sigma is to reduce process variation and improve quality.

Example:
Suppose you are manufacturing a product, and you want to achieve Six Sigma quality. This means that the process should be so precise that only 3.4 defects occur per million opportunities. If your process is currently producing 10 defects per million opportunities, you would need to improve it to meet the Six Sigma standard.

# Question 2: What type of data does not have a log-normal distribution or a Gaussian distribution? Give a proper example.

One example of data that does not have a log-normal or Gaussian distribution is count data, particularly data that follows a Poisson distribution. Poisson-distributed data represents the number of events occurring in fixed intervals of time or space.

Example:
The number of customer arrivals at a service center in a given hour follows a Poisson distribution. If you collect data on the number of arrivals per hour over several days, you may find that the distribution is not log-normal or Gaussian but rather follows the Poisson distribution.

# Question 3: What is the meaning of the five-number summary in Statistics? Give a proper example.

The five-number summary is a set of descriptive statistics that provides a concise summary of the distribution of a dataset. The five numbers include the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

Example:
Consider the dataset: 2, 7, 12, 15, 20, 22, 28, 35, 40.

The five-number summary would be:

Minimum: 2
Q1 (First Quartile): 12
Median (Q2): 20
Q3 (Third Quartile): 28
Maximum: 40

# Question 4: What is correlation? Give an example with a dataset & graphical representation on Jupyter Notebook.

Correlation measures the strength and direction of a linear relationship between two variables. It is represented by a correlation coefficient, usually denoted by 'r.' The value of 'r' ranges from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no correlation.

Example:
Let's consider a dataset of students' study hours and their exam scores:

```python
import numpy as np
import matplotlib.pyplot as plt

study_hours = np.array([2, 5, 7, 8, 10, 12, 15, 18])
exam_scores = np.array([50, 55, 60, 65, 70, 75, 80, 85])

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(study_hours, exam_scores)[0, 1]

# Scatter plot
plt.scatter(study_hours, exam_scores)
plt.title(f'Correlation: {correlation_coefficient:.2f}')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.show()
```
